**Desafio que envolve os dados utilizados no DataFrame 'dados_eleitorais' com algumas modificações, com o intuito de desenvolver um gráfico de pizza.**

**OBS: Esses dados são fictícios, isento de apoio à qualquer partido político, somente uma forma de colocar em prática alguns conceitos.**

```python
# Gráfico de pizza que representa a porcentagem de votos para cada candidato à presidência,
# com base na soma de votos, retirados do DataFrame 'dados_eleitorais'

# tirar a porcentagem de cada candidato

# total -> 1962.5 (100%) obs: avaliado em milhões

# Lula -> 35%
# Bolsonaro -> 58%
# Simone -> 5%
# Ciro -> 2 %

# importar a biblioteca 'matplotlib'

import matplotlib.pyplot as plt

# dados

porcentagem = [35,58,5,2] #Porcentagem de cada candidato, conforme a soma de votos.

cores = ['salmon','violet','cyan','pink'] # As cores que eu quero para representar cada fatia/candidato.

candidatos = ['Lula (PT)',' Jair Bolsonaro (PL)','Simone Tebet (PSDB)','Ciro Gomes (PDT)']

#Criar o gráfico

plt.pie (porcentagem,colors = cores, autopct='%1.0f%%', startangle = 140)

# Criar legenda

plt.legend(candidatos,loc='best')

# Criar título

plt.title('Porcentagem de votos (UF)')

#Exibir gráfico

plt.show()



```

**Gráfico de Pizza:**

```python
# Importando a biblioteca 'Matplotlip'

import matplotlib.pyplot as plt

# Dados

porcentagem = [40,30,15,10,5]  # representa cada modalidade esportiva na forma percentual;

cores = ['Green','purple','red','lightblue','yellow'] # cores de cada fatia;

rotulos = ['Futebol','Vôlei','Basquete','Natação','Outros'] # modalidades esportivas a ser avaliadas;


# Criar um gráfico de pizza

plt.pie(porcentagem, autopct='%1.1f%%', colors = cores, startangle = 140)

#startangle -> posição das fatias, conforme o ângulo;
#autopct -> formatação da str/porcentagem;

#Legendas

plt.legend(rotulos,loc='best')
#aceita dois argumentos: a lista de rótulos das fatias e a posição da legenda (loc);
#'best'-> escolhe automaticamente a melhor posição para colocar a legenda;

# Adicionar título
plt.title('Preferência por modalidades esportivas')

# Mostrar o gráfico
plt.show()
```
